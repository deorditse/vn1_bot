# Docx → Markdown Конвертер в Jupyter

Загрузи .docx → нажми кнопку → получи чистый Markdown с картинками и таблицами  
Под капотом: Pandoc + GFM (GitHub Flavored Markdown)


```python
import base64
import subprocess
import tempfile
import os


def docx_base64_to_markdown(docx_base64: str) -> str:
    # 1. Создаём временную директорию
    with tempfile.TemporaryDirectory() as tmpdir:
        docx_path = os.path.join(tmpdir, "input.docx")
        md_path = os.path.join(tmpdir, "output.md")

        # 2. Декодируем base64 → .docx
        with open(docx_path, "wb") as f:
            f.write(base64.b64decode(docx_base64))

        # 3. DOCX → Markdown (pandoc)
        subprocess.run(
            ["pandoc", docx_path, "-t", "markdown", "-o", md_path],
            check=True
        )

        # 4. Читаем markdown
        with open(md_path, "r", encoding="utf-8") as f:
            markdown = f.read()

        # 5. Файлы будут автоматически удалены при выходе из блока

    return markdown
```

```python

def test():
    with open("doc_example.docx", "rb") as f:
        base64_docx = base64.b64encode(f.read()).decode()

    markdown_text = docx_base64_to_markdown(base64_docx)

    print(markdown_text)


test()
```
