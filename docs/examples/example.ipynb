# Docx → Markdown Конвертер в Jupyter

Загрузи .docx → нажми кнопку → получи чистый Markdown с картинками и таблицами  
Под капотом: Pandoc + GFM (GitHub Flavored Markdown)


```python
import subprocess
import tempfile
import os

def convert(file_bytes: bytes) -> str:
    with tempfile.TemporaryDirectory() as tmpdir:
        docx_path = os.path.join(tmpdir, "input.docx")
        md_path = os.path.join(tmpdir, "output.md")
        # Записываем DOCX на диск
        with open(docx_path, "wb") as f:
            f.write(file_bytes)
        # pandoc → markdown
        subprocess.run(
            ["pandoc", docx_path, "-t", "markdown", "--wrap=preserve", "-o", md_path],
            check=True,       
            text=True,
        )
        # Читаем markdown
        with open(md_path, "r", encoding="utf-8") as f:
            return f.read()
```

```python
def test_md():
    with open("doc_example.docx", "rb") as f:
        file_bytes =  f.read()
    markdown_text = convert(file_bytes)
    print(markdown_text)


test_md()
```
