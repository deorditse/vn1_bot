```python
from langsmith import Client, traceable  # noqa: E402
from langgraph.prebuilt import create_react_agent  # noqa: E402


client = Client()

def get_weather(city: str) -> str:
    """Get weather for a given city."""
    return f"It's always sunny in {city}!"


# === Агент LangGraph ===
agent = create_react_agent(
    model=llm,
    tools=[get_weather],
    prompt="Ты — вежливый ассистент, который отвечает кратко.",
)

# === Оборачиваем в traceable ===
@traceable(run_type="chain", name="LangGraph GigaChat Agent")
def run_agent():
    return agent.invoke(-
        {"messages": [{"role": "user", "content": "What is the weather in Moscow?"}]}
    )
    

print(run_agent())
```
